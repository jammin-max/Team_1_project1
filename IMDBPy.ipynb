```python
#reading in netflix dataset
netflix = pd.read_csv('netflix_titles.csv')
netflix.info()
```

```python
#Creating a list of netflix titles to be used in for loop to return ratings
netflix_list = netflix['title'].tolist()
```

```python
#Created 4 sub lists of the movie names because for loop was timing out
list_1 = netflix_list[0:2000]
list_2 = netflix_list[2001:4000]
list_3 = netflix_list[4001:6000]
list_4 = netflix_list[6001:7787]
```

```python
#Empty lists to store the title, imdb ID, netflix title returned by the for loop for sublist 1.

title_list = []
imdbID = []
netflix_title = []
imdb_rating = []
total_votes = []
male_total_votes = []
male_rating = []
female_total_votes = []
female_rating = []
rating_demographics = []
count = 0

#looping through netflix title list to get the ratings for each title.
for title in list_1:
    
    #try to retrive a result for the title and get the rating details.
    #if no result is found skip down to except and print "movie not found".
    try:
        movies = ia.search_movie(str(title))
        title_rating = ia.get_movie(movies[0].movieID, 'vote details')
        
        #check to make sure the ratings detail is not blank.
        #If the rating details is blank, skip down to the except and print "no ratings data found"
        if title_rating.get('demographics') != None:
            
            #if the rating details is not blank store the rating details, title and imdb ID in a list.
            rating_demographics.append(title_rating.get('demographics'))
            netflix_title.append(title)
            imdbID.append(movies[0].movieID)
            
            #try to store the rest of the rating details in lists. If no data exists, store a "NaN value."
            try:
                imdb_rating.append(rating_demographics[count]['imdb users']['rating'])
            except KeyError:
                imdb_rating.append("NaN")
            try:
                total_votes.append(rating_demographics[count]['imdb users']['votes'])
            except KeyError:  
                total_votes.append("NaN")
            try:
                male_total_votes.append(rating_demographics[count]['males']['votes'])
            except KeyError:
                male_total_votes.append("NaN")
            try:
                male_rating.append(rating_demographics[count]['males']['rating'])
            except KeyError:
                male_rating.append("NaN")
            try:
                female_total_votes.append(rating_demographics[count]['females']['votes'])
            except KeyError:
                female_total_votes.append("NaN")
            try:
                female_rating.append(rating_demographics[count]['females']['rating'])
            except KeyError:
                female_rating.append("NaN")
            #increase the count by 1 so the apprpriate index can be used when retreiving the rating details.
            count+=1
        else:
            print(f"No ratings data found for {movies[0]} ID {movies[0].movieID}")
    except IndexError:
        print(f"{title} not found")
```

```python
#combining the results of the 3 for loops into one list
all_titles = title_list + title_list2 + title_list3 + title_list4
all_imdb_id = imdbID+imdbID2+imdbID3+imdbID4
all_netflix_titles = netflix_title+netflix_title2+netflix_title3+netflix_title4
all_imdb_ratings= imdb_rating+imdb_rating2+imdb_rating3+imdb_rating4
all_total_votes = total_votes+total_votes2+total_votes3+total_votes4
all_male_total_votes = male_total_votes+ male_total_votes2+male_total_votes3+male_total_votes4
all_male_rating = male_rating+male_rating2+male_rating3+male_rating4
all_female_total_votes = female_total_votes+female_total_votes2+female_total_votes3+female_total_votes4
all_female_rating = female_rating+female_rating2+female_rating3+female_rating4
```

```python
#converting list into a dict then into a df
netflix_rating_dict = {
    "title": all_netflix_titles,
    "imdb_title_id": all_imdb_id,
    "overall imdb rating": all_imdb_ratings,
    "total overall votes": all_total_votes,
    "male imdb rating": all_male_rating,
    "male votes": all_male_total_votes,
    "female imdb rating": all_female_rating,
    "female votes":all_female_total_votes
}

movie_rating_df = pd.DataFrame(netflix_rating_dict)
movie_rating_df.head()
```

```python
#merging the original netflix dataset with the movie ratings df
netflix_ratings = pd.merge(movie_rating_df,netflix, on='title', how='left')
```

```python
#Sorting df by highest rating
netflix_ratings = netflix_ratings.sort_values(by='overall imdb rating', ascending=False)
```

```python
#resetting the df index
netflix_ratings.reset_index(drop=True, inplace=True)
```

```python
#saving df to csv
netflix_ratings.to_csv('netflix_imdb_ratings.csv')
```
